Sea $x(t) = x^*$ con $x^* \in \mathbb{R}$ una solución de equilibrio de la ecuación
$$
\dot{x} = f(x).
$$

Entonces
$$
\dot{x} = \frac{d}{dt}x^* = 0,
$$
y por lo tanto
$$
f(x(t)) = f(x^*) = 0.
$$

Supongamos que existe una solución de la E.D.O. de la forma
$$
x(t) = x^* + \varepsilon \tilde{x}(t),
$$
para $\varepsilon \approx 0$.

Como $x(t)$ es solución, cumple
$$
\dot{x}(t) = f(x(t)).
$$

Así,
$$
\frac{d}{dt}\big(x^* + \varepsilon \tilde{x}(t)\big)
= f\big(x^* + \varepsilon \tilde{x}(t)\big),
$$
es decir,
$$
\varepsilon \dot{\tilde{x}}(t)
= f\big(x^* + \varepsilon \tilde{x}(t)\big).
$$

Tomamos la serie de Taylor de $f$ alrededor de $x^*$:
$$
\begin{aligned}
f\big(x^* + \varepsilon \tilde{x}(t)\big)
&= f(x^*)
+ f'(x^*)\big(x^* + \varepsilon \tilde{x}(t) - x^*\big) \\
&\quad + \frac{1}{2!} f''(x^*)\big(x^* + \varepsilon \tilde{x}(t) - x^*\big)^2
+ \cdots
\end{aligned}
$$

Simplificando:
$$
\begin{aligned}
f\big(x^* + \varepsilon \tilde{x}(t)\big)
&= f(x^*)
+ f'(x^*)\,\varepsilon \tilde{x}(t)
+ \frac{1}{2!} f''(x^*)\,\varepsilon^2 \tilde{x}^2(t)
+ \cdots \\
&= f'(x^*)\,\varepsilon \tilde{x}(t) + \mathcal{O}(\varepsilon^2),

= \lambda\,\varepsilon\,\tilde{x}(t) +  \mathcal{O}(\varepsilon^2)
\end{aligned}
$$

Así, tenemos que

$$
\varepsilon \frac{d}{dt}\tilde{x}(t)
= \lambda\,\varepsilon\,\tilde{x}(t)
+ \mathcal{O}(\varepsilon^2).
$$

$$
\frac{d}{dt}\tilde{x}(t) = \lambda \tilde{x}(t) + \mathcal{O}(\varepsilon),
$$

y haciendo $\varepsilon \to 0$ obtenemos

$$
\frac{d}{dt}\tilde{x}(t) = \lambda \tilde{x}(t).
$$

La solución es

$$
\tilde{x}(t) = C_0 e^{\lambda t}.
$$

- Si $\lambda > 0$, la solución **diverge**.
- Si $\lambda < 0$, la solución **converge a $0$**.
- Si $\lambda = 0$, la solución **no converge ni diverge**.

Por lo tanto, la solución de equilibrio

$$
x(t) = x^*
$$

es **estable** si

$$
\left.\frac{df}{dx}\right|_{x^*} < 0,
$$

e **inestable** si

$$
\left.\frac{df}{dx}\right|_{x^*} > 0.
$$

Más aún, si

$$
\left.\frac{df}{dx}\right|_{x^*} = 0,
$$

no podemos determinar si el equilibrio es estable o inestable mediante la linealización.

A este procedimiento se le llamara **Analisis de estabilida lineal**


### Ejemplo

Sea
$$
\dot{x} = \sin x.
$$

Los puntos de equilibrio satisfacen
$$
\sin x = 0,
$$
es decir,
$$
x = k\pi, \quad k \in \mathbb{Z}.
$$

Se observa que:

- Si
  $$
  x = 2k\pi,
  $$
  el equilibrio es **inestable**.

- Si
  $$
  x = (2k+1)\pi,
  $$
  el equilibrio es **estable**.

---

Ahora, usando el **análisis de estabilidad lineal (A.E.L.)**:

Como
$$
f(x) = \sin x,
$$
tenemos
$$
\frac{df}{dx} = f'(x) = \cos x.
$$

Evaluando en los puntos de equilibrio:
$$
f'(k\pi) = \cos(k\pi).
$$

Por lo tanto,

$$
\cos(n\pi) =
\begin{cases}
1, & \text{si } n \text{ es par}, \\
-1, & \text{si } n \text{ es impar}.
\end{cases}
$$

- Si $k$ es par, entonces $f'(x^*) > 0$ y el equilibrio es **inestable**.
- Si $k$ es impar, entonces $f'(x^*) < 0$ y el equilibrio es **estable**.


### Ejemplo

Sea
$$
\dot{x} = x^3,
$$
es decir,
$$
f(x) = x^3.
$$

El punto
$$
x^* = 0
$$
es un **punto de equilibrio** y es **estable** (en el sentido no lineal).

Ahora, aplicando el **análisis de estabilidad lineal (A.E.L.)**:

$$
f'(x) = 3x^2.
$$

Evaluando en el equilibrio:
$$
f'(0) = 0.
$$

En efecto, el análisis de estabilidad lineal **no nos dice nada** en este caso, ya que la derivada en el punto de equilibrio es cero.


### Ejemplo

Sea
$$
\dot{x} = -x^3.
$$

El punto
$$
x^* = 0
$$
es una **solución de equilibrio estable**.

Ahora, aplicando el **análisis de estabilidad lineal (A.E.L.)**:

$$
f(x) = -x^3,
\qquad
f'(x) = -3x^2.
$$

Evaluando en el equilibrio:
$$
f'(0) = 0.
$$

Por lo tanto, el análisis de estabilidad lineal **tampoco nos dice nada** en este caso.


### Ejemplo

Sea
$$
\dot{x} = x^2,
$$
de modo que
$$
f(x) = x^2.
$$

El punto
$$
x^* = 0
$$
es un **punto semiestable**.

Aplicando el **análisis de estabilidad lineal (A.E.L.)**:

$$
f'(x) = 2x.
$$

Evaluando en el equilibrio:
$$
f'(0) = 0.
$$

Por lo tanto, el análisis de estabilidad lineal **no permite concluir** la estabilidad del equilibrio.
