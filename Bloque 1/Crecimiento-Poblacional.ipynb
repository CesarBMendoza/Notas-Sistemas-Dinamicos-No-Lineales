### Modelo de hormigas

Un modelo de hormigas considera la colonización de nuevos sitios de alimentación
dado un camino de feromonas dejado por hormigas exploradoras.

Si $x(t)$ representa el número de hormigas alimentándose en el sitio dado:

- $\alpha > 0$ es la tasa a la cual las hormigas encuentran un nuevo sitio de alimentación.
- $\beta > 0$ es la tasa con la que las hormigas son atraídas al sitio mediante feromonas.
- $\eta > 0$ es el tamaño de la colonia.
- $S > 0$ es la tasa máxima con la cual abandonan el sitio de alimentación.
- $k > 0$ es el número de hormigas necesarias para que el sitio pierda hormigas a una tasa de $\dfrac{S}{2}$.

Entonces, el modelo está dado por la ecuación

$$
\dot{x} = (\alpha + \beta x)(\eta - x) - \frac{Sx}{x + k}.
$$

Ahora, Escribamos el modelo de manera no dimensional,
reescalando a $x$ y $t$ apropiadamente para reducir
el número de parámetros adimensionales.

Tomemos primero

$$
y = \frac{x}{\eta},
$$

por lo que

$$
\frac{dy}{dt} = \frac{1}{\eta}\frac{dx}{dt}.
$$

Además, tomemos el reescalamiento temporal

$$
\tau = \alpha t,
$$

de donde

$$
\frac{d\tau}{dt} = \alpha.
$$

Por lo tanto,

$$
\frac{dy}{dt}
= \frac{dy}{d\tau}\frac{d\tau}{dt}
= \alpha \frac{dy}{d\tau}.
$$

Como el modelo original es

$$
\dot{x} = (\alpha + \beta x)(\eta - x) - \frac{Sx}{x + k},
$$

entonces

$$
\eta \frac{dy}{dt}
= (\alpha + \beta \eta y)(\eta - \eta y)
- \frac{S \eta y}{\eta y + k}.
$$

Dividiendo entre $\eta$ y usando el reescalamiento temporal, obtenemos

$$
\eta \alpha \frac{dy}{d\tau}
= (\alpha + \beta \eta y)\eta(1 - y)
- \frac{S\eta y}{\eta( y + \frac{k}{\eta})}
$$
Es decir

$$
\alpha \frac{dy}{d\tau}
= (\alpha + \beta \eta y)(1 - y)
- \frac{S y}{\eta(y + \frac{k}{\eta})}
$$

Luego,

$$
\frac{dy}{dt}
= (\alpha - \beta \eta y)(1 - y)
- \frac{Sy}{\eta\left(y + \frac{k}{\eta}\right)}.
$$

$$
\frac{dy}{d\tau}
= (1 - \tfrac{\beta \eta}{\alpha} y)(1 - y)
- \frac{S}{\alpha \eta}\,\frac{y}{y + \tfrac{k}{\eta}}.
$$

Así, tomando los parámetros adimensionales

$$
a = \frac{\beta \eta}{\alpha}, \qquad
b = \frac{S}{\alpha \eta}, \qquad
c = \frac{k}{\eta},
$$

obtenemos finalmente el sistema no dimensional

$$
\frac{dy}{d\tau}
= (1 - a y)(1 - y) - \frac{b y}{y + c}.
$$

Los parámetros adimensionales tienen la siguiente interpretación:

$$
a = \frac{\beta n}{\alpha}
$$

representa el **número de hormigas que son atraídas a un nuevo sitio**
mediante feromonas.

$$
b = \frac{S}{n\alpha}
$$

representa la **tasa a la cual la colonia deja el viejo sitio**
para encontrar un nuevo sitio.

$$
c = \frac{k}{n}
$$

representa el **número “realmente” necesario de hormigas**,
es decir, **el número crítico**.
