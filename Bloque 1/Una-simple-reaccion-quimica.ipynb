# Una Simple Reaccion Quimica

Para comprender qué produce la ley de acción de masas al describir esta reacción simple,
completamos el modelo suponiendo que la reacción es reversible; es decir, que el producto
C se descompone espontáneamente de nuevo en los reactivos A y B a una tasa
$k_C$ proporcional a la cantidad de C. Esto da lugar a la reacción química

$$
A + B \;\xrightleftharpoons[\beta]{\alpha}\; C,
$$

y a las ecuaciones diferenciales

$$
\frac{dA}{dt} = -\alpha AB + \beta C,
$$

$$
\frac{dB}{dt} = -\alpha AB + \beta C,
$$

$$
\frac{dC}{dt} = \alpha AB - \beta C.
$$

Como

$$
\frac{dA}{dt} + \frac{dC}{dt} = 0,
$$

podemos deducir que A + C debe ser una constante; llamémosla $K_A$, de modo que

$$
A(t) + C(t) = A(0) + C(0) = K_A
$$

y podemos escribir A en función de C:

$$
A(t) = K_A - C(t).
$$

De manera análoga, tenemos que

$$
B(t) + C(t) = B(0) + C(0) = K_B.
$$

Es decir
$$
B(t) = K_B - c(t)
$$

Por lo tanto:

$$
\frac{dC}{dt} = \alpha (K_A - C)(K_B - C) - \beta C 
$$

y si encontramos los puntos de equilibrio de la sustancia $C$, encontramos los púntos de equilibrio de las sustancias $A$ y $B$

Asi, buscando las soluciones de equilibrio de C, tenemos que:

$$
0 = \alpha (K_A - C^*)(K_B - C^*) - \beta C^*\quad \text{Es decir}
$$

$$
0 = \alpha\, {C^*}^2 - \big[\alpha (K_A + K_B) + \beta \big]\, C^*
+ \alpha K_A K_B
$$

y por la formula general, tenemos que:

$$
C_{\pm}^* =
\frac{(K_A + K_B + K_d)\pm\sqrt{(K_A + K_B + K_d)^2 - 4K_AK_B}}{2}\quad \text{donde}\quad K_d = \frac{\beta}{\alpha}
$$

$K_d$ es llamada la **constante de disociación**.

Pero, notemos que el discriminante de nustras soluciones son siempre positivas, es decir:

$$
(K_A + K_B + K_d)^2 - 4K_AK_B =
$$

$$
= K_A^2 + K_B^2 + K_d^2
+ 2K_AK_B + 2K_AK_d + 2K_BK_d
- 4K_AK_B =
$$

$$
= (K_A^2 - 2K_AK_B + K_B^2)
+ K_d^2 + 2K_AK_d + 2K_BK_d =
$$

$$
= (K_A - K_B)^2 + K_d^2 + 2K_AK_d + 2K_BK_d > 0
$$

De donde

$$
\sqrt{(K_A + K_B + K_d)^2 - 4K_AK_B}\leq K_A + K_B + K_d
$$

Por lo cual

$$
0 < \frac{K_A + K_B + K_d - \sqrt{(K_A + K_B + K_d)^2 - 4K_AK_B}}{2} < \frac{K_A + K_B + K_d + \sqrt{(K_A + K_B + K_d)^2 - 4K_AK_B}}{2}
$$

Por lo tanto,
$$
0 < C_{-}^* < C_{+}^*
$$

Ahora queremos determinar como es la estabilidad de cada solucion

pero notemos que:

$$
\frac{dC}{dt} =  \alpha {C^*}^2 - \big[\alpha (K_A + K_B) + \beta \big]\, C^*
+ \alpha K_A K_B
$$
y esto nos dice que es una parabola que abre hacia arriba, por lo que

$C_{-}^*$ Es **Estable**

$C_{+}^*$ Es **Inestable**

Otra forma de ver esto es viendo cuando $f'(C)<0$ donde:

$$
f(C) = \alpha {C^*}^2 - \big[\alpha (K_A + K_B) + \beta \big] C^* + \alpha K_A K_B \quad \text{Asi}
$$

$$
f'(C) = \alpha \big[2C - (K_A + K_B)\big] - \beta
$$

por lo tanto, buscamos cuando

$$
0 > \alpha \big[2C - (K_A + K_B)\big] - \beta
$$

Es decir

$$
0 > 2C - (K_A + K_B) - K_d
$$

Asi

$$
C < \frac{K_A + K_B + K_d}{2} \quad \text {y esto solo ocurre para $C_{-}^*$}
$$

Es deccir $f'(C_{-}^*)<0$ (Asi en efecto $C_{-}^*$ es **Estable**)

Esto parece sugerir que existe una región en el espacio de fases para la cual la cantidad
del químico $C$ crece sin límite; sin embargo, esto es ilusorio, ya que los equilibrios
$C_\pm^*$ están definidos en términos de la condición inicial (presumiblemente conocida)
$C(0)$, y podemos mostrar que $C(0)$ está muy por debajo del nivel umbral $C_+^*$, Pues:

$$
\frac{C_-^* + C_+^*}{2}
= \frac{1}{2}\,(K_A + K_B + K_d)
= C(0) + \frac{1}{2}\,(A(0) + B(0) + K_d)
> C(0).
$$

Es decir, $C(0)$ está por debajo del punto medio entre $C_-^*$ y $C_+^*$

Por lo tanto, $C_-^*$ es globalmente asintóticamente estable, y el nivel del químico
$C$ se acercará a $C_-^*$ (a partir del cual también podemos calcular las soluciones de
equilibrio de $A$ y $B$)

Ademas, Notemos que como $C_{-}^*$, $A_{-}^*$ y $B_{-}^*$ son soluciones de equilibrio del sistema, en particular

$$
0 = \alpha A_{-}^* B_{-}^* - \beta C_{-}^* \quad i.e.
$$

$$
\frac{\beta}{\alpha} = \frac{A_{-}^* B_{-}^*}{C_{-}^*} \quad \text{Asi}\quad K_d = \frac{A_{-}^* B_{-}^*}{C_{-}^*}
$$
