Consideremos el sistema en coordenadas polares

$$
\begin{cases}
\dot r = r(1 - r^2), \\
\dot \theta = \mu - \sin\theta.
\end{cases}
$$

Las soluciones radiales satisfacen:

$$
r = 0 \quad \text{y} \quad r = 1.
$$

Calculamos la derivada de $\dot r$ respecto de $r$:

$$
\frac{d}{dr}\big(r(1 - r^2)\big) = 1 - 3r^2.
$$

Evaluando en los puntos críticos:

- En $r = 0$:
  $$
  |1 - 3r^2|_{r=0} = 1,
  $$
  por lo que el equilibrio es **inestable**.

- En $r = 1$:
  $$
  |1 - 3r^2|_{r=1} = 2 > 1,
  $$
  por lo que el equilibrio es **estable**.

Por lo tanto, el sistema presenta un **ciclo límite estable** en $r = 1$.
