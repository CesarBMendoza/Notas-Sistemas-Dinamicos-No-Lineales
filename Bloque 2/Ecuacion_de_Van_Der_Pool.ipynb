### Ecuación de Van der Pol

La ecuación de Van der Pol está dada por

$$
\ddot{x} + \mu (x^2 - 1)\dot{x} + x = 0,
$$

donde $\mu > 0$.

Identificando las funciones del sistema de Liénard, tenemos:

$$
g(x) = x,
$$

$$
f(x) = \mu (x^2 - 1).
$$

Además,

$$
F(x) = \int_0^x (u^2 - 1)\,du
$$

entonces

$$
F(x) = \left[\frac{u^3}{3} - u\right]_0^x
     = \frac{x^3}{3} - x.
$$

Para encontrar los ceros de $F(x)$ resolvemos

$$
\frac{x^3}{3} - x = 0,
$$

lo que equivale a

$$
x(x^2 - 3) = 0.
$$

Por tanto,

$$
x = 0 \quad \text{o} \quad x_0 = \pm \sqrt{3}.
$$

En particular, para $x_0 = \sqrt{3}$ se cumple que

$$
F(x_0) = 0.
$$

Ahora, haciendo

$$
\ddot{x} + \mu (x^2 - 1)\dot{x} = -x.
$$

Observamos que

$$
\left(\dot{x} + \mu(\frac{x^3}{3} - x)\right)' = -x,
$$

y por lo tanto

$$
\frac{d}{dt}\left[\dot{x} + \mu F(x)\right] = -x,
$$

donde

$$
F(x) = \frac{x^3}{3} - x.
$$

Definimos la variable auxiliar

$$
w = \dot{x} + \mu F(x).
$$

Entonces

$$
\dot{w} = -x.
$$

Así, el sistema de segundo orden puede escribirse como el sistema equivalente de primer orden

$$
\begin{cases}
\dot{x} = w - \mu F(x), \\
\dot{w} = -x.
\end{cases}
$$

Si además definimos

$$
y = \dfrac{w}{\mu},
$$

entonces

$$
\dot{y} = \dfrac{\dot{w}}{\mu} = -\dfrac{x}{\mu}.
$$


(a) Argumenta porqué el periodo de oscilación $T$ del ciclo límite se puede aproximar por

$$
T \approx 2 \int_{x_a}^{x_{\min}} \frac{db}{\dot{x}}\, dx
$$

donde $x_a$ es la solución a la ecuación algebraica $f(x_a)=f_{\max}$, donde $f_{\max}$ es el valor del máximo local de $f(x)$, y
$x_a \ne x_{\max}$, i.e. no es la coordenada horizontal del máximo local de $f(x)$,
y $x_{\min}$ es la coordenada horizontal del mínimo local de $f(x)$.

$$
\longrightarrow
$$

Como ya sabemos que hay un periodo de oscilación y si hacemos a $t$ una función que dependa de $x$, tendríamos


además, notemos que es: 

$$
\ddot{x} + \mu (x^2 - 1)\dot{x} + x = 0.
$$

Si tomamos

$$
z=-x,\qquad \dot{z}=-\dot{x},\qquad \ddot{z}=-\ddot{x}
$$

por tanto:

$$
-\ddot{z} - \mu (z^2-1)\dot{z} - z = 0
\quad \Leftrightarrow \quad
\ddot{z} + \mu (z^2-1)\dot{z} + z = 0.
$$

Así, esto nos dice que nuestro sistema es simétrico, por lo que el periodo de oscilación es muy cercano a lo que tarde $2$ veces cuando subimos del mínimo al máximo.

$$
T \approx 2\bigl(t(x_a)-t(x_{\min})\bigr)
= 2\int_{x_a}^{x_{\min}} \frac{dt}{dx}\,dx.
$$

Así,

$$
T \approx 2\int_{x_a}^{x_{\min}} \frac{dt}{dx}\,dx.
$$

(b) Argumenta porqué la ec. algebraica de $x_a$

$$
F(x_a)=\overline{F}_{\max}
$$

puede escribirse como

$$
(x-x_{\max})^2(x-x_a)=0
$$

y encuentra el valor de $x_a$.

Calculemos $\overline{F}_{\max}$. Así como

$$
F(x)=\frac{1}{3}x^3-x.
$$

$$
\Rightarrow F'(x)=x^2-1
$$

y haciendo $0=F'(x)=(x-1)(x+1)$

y como $F''(x)=2x$ y $F''(-1)=-2$, se tiene

$$
x_{\max}=-1
$$

pero

$$
F(-1)=-\frac{1}{3}+1=\frac{2}{3}.
$$

y

$$
\overline{F}_{\max}=\frac{2}{3}.
$$

Así, ahora resolviendo

$$
\frac{1}{3}x^3 - x = \frac{2}{3}
$$

o.e.

$$
x^3 - 3x = 2,
$$

y como tenemos de hint que

$$
(x-x_{\max})^2(x-x_a)=0
$$

y $x_{\max}=-1$.

Así,

$$
(x+1)^2 = x^2 + 2x + 1.
$$

Así tenemos que:

En efecto, la ecuación

$$
F(x)=\overline{F}_{\max}
$$

la podemos reescribir como

$$
(x+1)^2(x-2)=0.
$$

Donde además, $x_a=2 \ne -1=x_{\max}$.

(c) Usando que $\dot{y}=F(x)$ y la ecuación de $\dot{y}$, muestra que:

$$
\frac{dt}{dx} = -\frac{\mu(x^2-1)}{x}.
$$

Como $y \approx F(x)$, supongamos que $y = F(x)$, por tanto

$$
\dot{y} = F'(x)\dot{x} = F'(x)\frac{dx}{dt},
$$

por tanto

$$
\frac{dt}{dx} = \frac{F'(x)}{\dot{y}} = \frac{x^2-1}{-\frac{1}{\mu}x}
= -\mu\frac{(x^2-1)}{x}.
$$


(d) Usando lo anterior, encuentra una expresión aproximada para $T$ en función del parámetro $\mu$

Como

$$
T \approx 2\int_{x_a}^{x_{\min}} \frac{dt}{dx}\,dx
= 2\int_{2}^{1} \frac{-\mu(x^2-1)}{x}\,dx
$$

$$
= -2\mu \int_{2}^{1} \left(x-\frac{1}{x}\right)\,dx
= 2\mu \int_{1}^{2} \left(x-\frac{1}{x}\right)\,dx
$$

$$
= 2\mu \left(\frac{x^2}{2}-\ln x\right)\Big|_{1}^{2}
= 2\mu \left(\frac{3}{2}-\ln(2)\right).
$$

y.e.

$$
T \approx 2\mu\left(\frac{3}{2}-\ln(2)\right).
$$

**2.-** En lo anterior consideramos la ec. de Van der Pol

$$
\ddot{x}+x+\mu(x^2-1)\dot{x}=0,
$$

con $\mu \gg 1$, i.e. en el límite altamente no-lineal. Ahora consideremos, con $\mu \ll 1$, i.e. en el límite débilmente no-lineal.

**(a)** Muestra que la solución general a la ecuación lineal que resulta cuando $\mu=0$ es

$$
x(t)=r\cos(t+\phi),
$$

con $r,\phi\in\mathbb{R}$

Como $\mu=0$, entonces

$$
\ddot{x}+x=0,
$$

y haciendo

$$
y=\dot{x},
$$

entonces

$$
\dot{y}=\ddot{x}=-x.
$$

Por tanto, tenemos:

$$
\begin{cases}
\dot{x}=y,\\
\dot{y}=-x
\end{cases}
\qquad \Rightarrow \qquad
\begin{pmatrix}
\dot{x}\\
\dot{y}
\end{pmatrix}
=
\begin{pmatrix}
0 & 1\\
-1 & 0
\end{pmatrix}
\begin{pmatrix}
x\\
y
\end{pmatrix}.
$$

Y como

$$
\begin{pmatrix}
0 & 1\\
-1 & 0
\end{pmatrix}
$$

es una M. de rotación, sabemos que las soluciones corresponden con:

$$
x(t)=r\cos(t+\theta)
$$

y

$$
y(t)=r\sin(t+\theta),
$$

con $r,\theta\in\mathbb{R}$

**(b)** Considera la ecuación lineal inhomogénea

$$
\ddot{x}+x=g(t),
$$

muestra que si

$$
g(t)=a\cos(t+b),
$$

con $a,b\in\mathbb{R}$, entonces las soluciones muestran el término divergente

$$
t\cos(t),
$$

pero si

$$
g(t)=a\cos(\omega t+b)
\quad \text{con} \quad \omega\neq 1,
$$

entonces las soluciones no tienen términos divergentes

**Hint:** usa $x(t)=At\cos(t+B)$ en el primer caso, y $x(t)=A\cos(\omega t+B)$ para el segundo caso.

**Caso 1)** Si

$$
g(t)=a\cos(t+b),
\qquad \text{con } a,b\in\mathbb{R},
$$

proponemos

$$
x(t)=At\cos(t+B),
$$

y

$$
\dot{x}=At\big(-\sin(t+B)\big)+A\cos(t+B).
$$

Entonces,

$$
\ddot{x}=At\big(-\cos(t+B)\big)-\sin(t+B)(A)-A\sin(t+B).
$$

Por tanto,

$$
\ddot{x}+x
=
At\cos(t+B)-At\cos(t+B)-2A\sin(t+B)
$$

$$
=-2A\sin(t+B)
=-2A\cos\left(t+B-\frac{\pi}{2}\right).
$$

Así,

$$
-2A\cos\left(t+B-\frac{\pi}{2}\right)=a\cos(t+b).
$$

Y en efecto, $x(t)=At\cos(t+B)$ tiene el término divergente $t\cos(t+B)$.

**Caso 2)** Si

$$
g(t)=a\cos(\omega t+b),
$$

y consideramos ahora

$$
x(t)=A\cos(\omega t+B),
$$

por tanto,

$$
\dot{x}=-A\omega\sin(\omega t+B),
$$

y

$$
\ddot{x}=-A\omega^2\cos(\omega t+B).
$$

$$
\ddot{x}+x=A\cos(\omega t+B)-A\omega^2\cos(\omega t+B)
$$

$$
= A(1-\omega^2)\cos(\omega t+B)=a\cos(\omega t+b).
$$

Así también, las soluciones

$$
x(t)=A\cos(\omega t+B)
$$

no tienen términos divergentes
