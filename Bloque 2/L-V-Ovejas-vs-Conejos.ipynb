# Sistema de competencia por recursos

Consideramos el sistema

$$
\begin{cases}
\dot x = x(3-x-2y),\\
\dot y = y(2-x-y).
\end{cases}
$$

Este es un caso particular del modelo de competencia de Lotka--Volterra en dos dimensiones:

$$
\begin{cases}
\dot x = x(r_1-a_{11}x-a_{12}y),\\
\dot y = y(r_2-a_{21}x-a_{22}y).
\end{cases}
$$

En este caso, los parámetros son:

$$
r_1=3,\quad r_2=2,\quad a_{11}=1,\quad a_{12}=2,\quad a_{21}=1,\quad a_{22}=1.
$$

Aquí, $x(t)$ e $y(t)$ representan dos poblaciones que compiten por recursos limitados.

## 1. Nulclinas

Las nulclinas se obtienen imponiendo $\dot x=0$ y $\dot y=0$.

### Nulclinas de $x$

$$
\dot x = x(3-x-2y)=0
$$

Entonces:

$$
x=0
\qquad \text{o} \qquad
3-x-2y=0.
$$

La segunda recta puede escribirse como

$$
y=\frac{3-x}{2}.
$$

### Nulclinas de $y$

$$
\dot y = y(2-x-y)=0
$$

Entonces:

$$
y=0
\qquad \text{o} \qquad
2-x-y=0.
$$

La segunda recta puede escribirse como

$$
y=2-x.
$$

## 2. Puntos de equilibrio

Los puntos de equilibrio son las intersecciones de las nulclinas.

### Sobre los ejes
- $(0,0)$
- $(3,0)$
- $(0,2)$

### Equilibrio interior
Buscamos la intersección de

$$
3-x-2y=0,
\qquad
2-x-y=0.
$$

Restando ambas ecuaciones:

$$
(3-x-2y)-(2-x-y)=0
\Rightarrow 1-y=0
\Rightarrow y=1.
$$

Sustituyendo en $2-x-y=0$:

$$
2-x-1=0
\Rightarrow x=1.
$$

Por lo tanto, aparece el equilibrio interior

$$
(1,1).
$$

## 3. Jacobiano

El jacobiano del sistema es

$$
J(x,y)=
\begin{pmatrix}
\frac{\partial \dot x}{\partial x} & \frac{\partial \dot x}{\partial y}\\[4pt]
\frac{\partial \dot y}{\partial x} & \frac{\partial \dot y}{\partial y}
\end{pmatrix}.
$$

Como

$$
\dot x = 3x-x^2-2xy,
\qquad
\dot y = 2y-xy-y^2,
$$

resulta

$$
J(x,y)=
\begin{pmatrix}
3-2x-2y & -2x\\
-y & 2-x-2y
\end{pmatrix}.
$$

## 4. Clasificación de los equilibrios

### En $(0,0)$

$$
J(0,0)=
\begin{pmatrix}
3 & 0\\
0 & 2
\end{pmatrix}.
$$

Los autovalores son $3$ y $2$, ambos positivos.  
Por tanto, $(0,0)$ es una **fuente inestable**.

### En $(3,0)$

$$
J(3,0)=
\begin{pmatrix}
-3 & -6\\
0 & -1
\end{pmatrix}.
$$

Los autovalores son $-3$ y $-1$, ambos negativos.  
Entonces $(3,0)$ es un **nodo estable**.

### En $(0,2)$

$$
J(0,2)=
\begin{pmatrix}
-1 & 0\\
-2 & -2
\end{pmatrix}.
$$

Los autovalores son $-1$ y $-2$, ambos negativos.  
Entonces $(0,2)$ es también un **nodo estable**.

### En $(1,1)$

$$
J(1,1)=
\begin{pmatrix}
-1 & -2\\
-1 & -1
\end{pmatrix}.
$$

Su polinomio característico es

$$
\det
\begin{pmatrix}
-1-\lambda & -2\\
-1 & -1-\lambda
\end{pmatrix}
=
(-1-\lambda)^2-2.
$$

Así,

$$
(\lambda+1)^2=2
\Rightarrow
\lambda=-1\pm \sqrt{2}.
$$

Uno de los autovalores es positivo y el otro negativo, por lo que $(1,1)$ es un **punto silla**.

## 5. Interpretación dinámica

Este sistema describe competencia entre dos especies.

- Si una trayectoria termina en $(3,0)$, la especie $x$ sobrevive y $y$ se extingue.
- Si una trayectoria termina en $(0,2)$, la especie $y$ sobrevive y $x$ se extingue.
- El equilibrio interior $(1,1)$ es inestable, así que no representa coexistencia estable.
- La variedad estable de la silla $(1,1)$ separa las condiciones iniciales que van hacia $(3,0)$ de las que van hacia $(0,2)$.

## 6. Conclusión

El sistema presenta **exclusión competitiva**: no hay coexistencia estable en el interior del primer cuadrante.  
Dependiendo de la condición inicial, una de las dos especies desplaza a la otra.

Los atractores del sistema en el primer cuadrante son:

$$
(3,0)
\qquad \text{y} \qquad
(0,2).
$$

El punto $(1,1)$ actúa como frontera entre ambas cuencas de atracción.

1. Tasa de crecimiento mayor en los conejos.

2. Los conflictos (interacciones) son proporcionales al tamaño de la población general.

3. Se aplica una sanción en las tasas de crecimiento cada vez que hay una interacción, por lo que los conejos sufren mayor penalización.

Un modelo que cumple los supuestos es:

$$
\dot{x} = x(3 - x - 2y) = 3x - x^2 - 2xy,
$$

$$
\dot{y} = y(2 - x - y) = 2y - xy - y^2.
$$ 
Donde: 

$y(t)$ = población de ovejas.  
$x(t)$ = población de conejos.

Notemos ahora que las soluciones de equilibrio son:

$$
x_1^*= (0,0), \quad x_2^* = (3,0), \quad x_3^* = (0,2).
$$

y estas ultimas 2 soluciones nos dicen que Si una no está, la otra llega a su capacidad de carga.

$$
x^* = (1,1).
$$

Recordemos que:

Si $x^*$ es un nodo, las soluciones son paralelas a las direcciones correspondientes al vector propio del valor propio.

Vemos que los tipos de puntos son, Es decir, veamos su estabilidad

Calculemos el jacobiano del sistema: 
$$
\dot{x} = x(3 - x - 2y) = 3x - x^2 - 2xy,
$$

$$
\dot{y} = y(2 - x - y) = 2y - xy - y^2.
$$  

El cual es: 

$$
J =
\begin{pmatrix}
3 - 2x - 2y & -2x \\
-y & 2 - x - 2y
\end{pmatrix}.
$$

Así,

$$
J_{(0,0)} =
\begin{pmatrix}
3 & 0 \\
0 & 2
\end{pmatrix}
\Rightarrow \operatorname{tr}(J_{(0,0)}) = 5>0 \quad y\quad \det(J_{(0,0)}) = 6> 0
$$ 

Por lo que veamos el discriminante, por lo que consideremos:

$$
\operatorname{tr}(J_{(0,0)})^2 - 4\det(J_{(0,0)}) = 25 - 24 > 0.
$$

Por lo tanto $(0,0)$ es un **nodo inestable**.

Sus valores propios son $\lambda = 2, 3$ y $V_1=(0,1)$ es su vector propío,

ahora, Para $(0,2)$,

$$
J_{(0,2)} =
\begin{pmatrix}
-1 & 0 \\
-2 & -2
\end{pmatrix}.
$$

Se tiene que

$$
\operatorname{tr}(J_{(0,2)}) = -3,
\quad
\det(J_{(0,2)}) = 2 \quad y \quad \operatorname{tr}(J_{(0,2)})^2 - 4\det(J_{(0,2)}) = 1.
$$

Por tanto $(0,2)$ es un nodo estable y cuyos valores propios son $-1, -2$,
y $(1,-2)$ es vector propio.

Por ultimo, Para $(3,0)$,

$$
J_{(3,0)} =
\begin{pmatrix}
-3 & -6 \\
0 & -1
\end{pmatrix}.
$$

Se tiene que

$$
\operatorname{tr}(J_{(3,0)}) = -4
\quad \text{y} \quad
\det(J_{(3,0)}) = 3.
$$

Por tanto $(3,0)$ es un **Nodo estable**

Y $\lambda = -1, -3$ son valores propios y $(3,-1)$ es el vector propio asociado a $-1$.



```python
import numpy as np
import matplotlib.pyplot as plt

def plot_clean_phase():
    x = np.linspace(-0.5, 3.5, 300)
    y = np.linspace(-0.5, 2.8, 300)
    X, Y = np.meshgrid(x, y)

    DX = X * (3 - X - 2*Y)
    DY = Y * (2 - X - Y)

    fig, ax = plt.subplots(figsize=(8, 6))

    ax.streamplot(
        X, Y, DX, DY,
        density=1.25,
        linewidth=1.8,
        arrowsize=1.0
    )

    xx = np.linspace(-0.5, 3.5, 800)

    ax.plot(np.zeros_like(xx), xx, color='purple', linewidth=2.2)
    ax.plot(xx, (3 - xx)/2, color='purple', linewidth=2.2)
    ax.plot(xx, np.zeros_like(xx), color='purple', linewidth=2.2)
    ax.plot(xx, 2 - xx, color='purple', linewidth=2.2)

    ax.plot([0, 3, 0, 1], [0, 0, 2, 1], 'o', color='black', markersize=5)

    ax.set_xlim(-0.5, 3.5)
    ax.set_ylim(-0.5, 2.8)
    ax.set_aspect('equal')
    ax.set_xticks([])
    ax.set_yticks([])

    for spine in ax.spines.values():
        spine.set_visible(False)

    plt.show()

plot_clean_phase()
```
