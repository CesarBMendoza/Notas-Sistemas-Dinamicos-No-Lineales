**Considere el sistema:**

$$
\begin{cases}
\dot{x} = x + e^{-y},\\
\dot{y} = -y
\end{cases}
$$

el cual tiene un punto silla en $(-1,0)$. Para aproximar su variedad estable, haz el cambio de variable $u=x+1$, y supón que la v.e. estable sigue la ecuación

$$
y = a_1u + a_2u^2 + O(u^3).
$$

Encontrando la expresión $\dfrac{dy}{du}$, determina los coeficientes $a_1$ y $a_2$ y comprueba con una gráfica que la curva $y(x)$ que obtuviste concuerda visualmente con la v.e. estable del sistema de ecuaciones dado su campo vectorial.

**Hint:**

$$
\frac{dy}{du}=\frac{dy/dt}{du/dt}
$$

y recuerda que al estar cerca de cierto punto, puedes aproximar funciones no polinomiales a partir de polinomios usando series de Taylor.

Primero hagamos el cambio de variable

$$
u=x+1
$$

Así,

$$
\frac{du}{dt}=\dot{u}=\dot{x}=x+e^{-y}=u-1+e^{-y}
$$

por lo cual

$$
\begin{cases}
\dot{u}=u-1+e^{-y},\\
\dot{y}=-y
\end{cases}
$$

y supongamos que la v. estable sigue la ecuación:

$$
y=a_1u+a_2u^2+O(u^3)
$$

Así, tenemos que

$$
\dot{y}=-y=-a_1u-a_2u^2-O(u^3),
$$

y como

$$
\frac{dy}{du}=\frac{dy/dt}{du/dt}=\frac{\dot{y}}{\dot{u}},
$$

pero como $\dot{u}=u-1+e^{-y}$, hagamos la serie de Taylor de $e^{-y}$ alrededor de $y=0$, i.e.

$$
e^{-y}=1-y+\frac{y^2}{2}+O(y^3).
$$

Que podríamos decir que

$$
e^{-y}\approx 1-y+\frac{y^2}{2}.
$$

por lo que:

$$
\dot{u}=u-1+e^{-y}=u-1+\left(1-y+\frac{y^2}{2}\right)
$$

$$
= u-y+\frac{y^2}{2},
$$

pero como

$$
y\approx a_1u+a_2u^2,
$$

entonces:

$$
\dot{u}=u-\left(a_1u+a_2u^2\right)+\frac{\left(a_1u+a_2u^2\right)^2}{2}
$$

$$
= u-a_1u-a_2u^2+\frac{a_1^2u^2+2a_1a_2u^3+a_2^2u^4}{2}
$$

$$
= (1-a_1)u+\left(-a_2+\frac{a_1^2}{2}\right)u^2+O(u^3)
$$

es decir,

$$
\dot{u}\approx (1-a_1)u+\left(-a_2+\frac{a_1^2}{2}\right)u^2.
$$

Por tanto:

$$
\frac{\dot{y}}{\dot{u}}
=
\frac{-a_1u-a_2u^2}{(1-a_1)u+\left(-a_2+\frac{a_1^2}{2}\right)u^2}.
$$

$$
\frac{-a_1-a_2u}{(1-a_1)+\left(-a_2+\frac{a_1^2}{2}\right)u}
$$

pero por otro lado, como

$$
y=a_1u+a_2u^2,
$$

entonces



Como

$$
\frac{dy}{du}
=
\frac{-a_1-a_2u}{(1-a_1)+\left(-a_2+\frac{a_1^2}{2}\right)u},
$$

Sea

$$
z=\frac{-a_1-a_2u}{c_1+c_2u},
$$

aproximémosla con Taylor en $u=0$ por lo que:

$$
z = z\big|_{u=0}+z'\big|_{u=0}\,u+O(u^2)
\approx -\frac{a_1}{c_1}
+
\frac{(-a_2)(c_1+c_2u)-c_2(-a_1-a_2u)}{(c_1+c_2u)^2}\Bigg|_{u=0}\,u=
$$

$$
= -\frac{a_1}{c_1}
+
\frac{-a_2c_1+a_1c_2}{c_1^2}\,u
$$

Pero como
$$
z=\frac{dy}{du}=a_1+2a_2u
$$

Por lo tanto,
$$
a_1+2a_2u=-\frac{a_1}{c_1}+\frac{-a_2c_1+a_1c_2}{c_1^2}\,u
$$

Así,

$$
a_1=-\frac{a_1}{c_1}
\qquad\text{y}\qquad
2a_2=\frac{-a_2c_1+a_1c_2}{c_1^2}
$$

i.e.

$$
a_1=-\frac{a_1}{1-a_1},
\qquad
2a_2=\frac{-a_2}{1-a_1}+\frac{a_1\left(-a_2+\frac{a_1^2}{2}\right)}{(1-a_1)^2}.
$$

$$
\Rightarrow a_1(1-a_1)=-a_1,
\qquad a_1=0.
$$

Si $a_1\neq 0$, entonces

$$
-a_1=-1,
\qquad a_1=2.
$$

Así, para la segunda ecuación, tenemos que para $a_1=0$,

$$
2a_2=\frac{0-a_2}{1}
\quad\Rightarrow\quad
3a_2=0,
$$

i.e.

$$
a_2=0.
$$

Y para $a_1=2$, tenemos que

$$
2a_2=a_2+2(-a_2+2),
$$

y si $a_2=\frac{4}{3}$,

$$
a_2=-2a_2+4
\quad\Rightarrow\quad
3a_2=4.
$$

Por lo tanto,

$$
a_2=\frac{4}{3}.
$$

Por lo tanto, la rama estable aproximada es

$$
y=2u+\frac{4}{3}u^2+O(u^3),
\qquad u=x+1.
$$

Equivalentemente,

$$
y=2(x+1)+\frac{4}{3}(x+1)^2+O\big((x+1)^3\big).
$$
