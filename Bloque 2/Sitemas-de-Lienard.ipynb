## Teorema (Liénard)

Un sistema de Liénard
$$
\ddot x + g(x) + f(x)\dot x = 0
$$

tiene un **único ciclo límite estable** si se cumplen las siguientes condiciones:

1. $f,g \in C^1$
2. $g(x)$ es impar.
3. $g(x) > 0$ para todo $x > 0$.
4. $f(x)$ es par.
5. $F(x) = \displaystyle \int_0^x f(s)\,ds$ es tal que: Existe un $x_0 > 0$
   - $F(x_0) = 0$,
   - $F(x) < 0$ para $0 < x < x_0$,
   - $F(x) > 0$ para $x > x_0$


## Ejemplo

### Oscilador armónico amortiguado

Consideremos el oscilador armónico amortiguado dado por

$$
m\ddot{x} = -kx - \gamma \dot{x}.
$$

Dividiendo entre $m$, obtenemos

$$
\ddot{x} + \frac{\gamma}{m}\dot{x} + \frac{k}{m}x = 0.
$$

Tomando la forma de Liénard, identificamos:

$$
g(x) = \frac{k}{m}x,
$$

$$
f(x) = \frac{\gamma}{m}.
$$
