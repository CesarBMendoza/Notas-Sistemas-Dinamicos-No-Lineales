### Bifurcación de silla–nodo de ciclos

**Ejemplo:**

$$
\begin{cases}
\dot r = \mu r + r^3 - r^5, \\
\dot \theta = \omega + \gamma .
\end{cases}
$$

---

### Soluciones de equilibrio

Las soluciones de equilibrio radiales satisfacen

$$
\dot r = 0 \quad \Longrightarrow \quad r = 0
$$

o bien

$$
0 = \mu + r^2 - r^4 .
$$

Esto equivale a

$$
r^4 - r^2 - \mu = 0.
$$

Resolviendo para $r^2$:

$$
r_{1,2}^2 = \frac{1 \pm \sqrt{1 + 4\mu}}{2}.
$$

Por lo tanto,

$$
r_{1,2,3,4} = \pm \sqrt{\frac{1 \pm \sqrt{1 + 4\mu}}{2}}.
$$

---

### Condiciones de existencia

Para que las soluciones existan se requiere

$$
1 + 4\mu \ge 0 \quad \Longrightarrow \quad \mu \ge -\frac{1}{4}.
$$

Además, debe cumplirse

$$
1 - \sqrt{1 + 4\mu} \ge 0,
$$

lo cual implica

$$
\sqrt{1 + 4\mu} \le 1
\quad \Longrightarrow \quad
-1 \le 1 + 4\mu \le 1.
$$
