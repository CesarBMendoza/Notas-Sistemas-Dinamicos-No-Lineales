Una población de un depredador y una presa  
puede modelarse mediante el sistema de  
ecuaciones en diferencias.

$$
R_{n+1} = (r+1)R_n - \gamma R_n^2 - bR_nF_n
$$

$$
F_{n+1} = cR_nF_n + (1-d)F_n
$$

Donde $r,b,c,d$ son parámetros positivos.

**(a)** ¿Qué pasa si se extingue la población $R_n$ o $F_n$?  
¿Quién es el depredador y quién es la presa?

$\to$ El depredador es $F_n$, y la presa es $R_n$.

$\to$ ¿Qué pasa si se extingue la población $R_n$ o $F_n$?

$\longrightarrow$ Si la población $F_n$ se extingue $(F_n=0)$,

entonces
$$
\begin{cases}
R_{n+1} = (r+1)R_n - \gamma R_n^2 \\
F_{n+1} = 0
\end{cases}
$$

tenemos que la presa crece
$$
R_{n+1} = (r+1)R_n - \gamma R_n^2
$$

y como crece de manera logística, crece  
hasta su capacidad de carga.

$\longrightarrow$ Si la población $R_n$ se extingue $(R_n=0)$,

tenemos que
$$
\begin{cases}
R_{n+1} = 0 \\
F_{n+1} = (1-d)F_n
\end{cases}
$$

y esto último nos dice que $F_n$ decrece exponencialmente.  
Si $1-d \leq 0$ se para, pero además

$$
1 > 1-d > 0 \iff 1 > d > 0
$$

**(b)** Encuentra todas las soluciones de equilibrio  
del sistema.

Como
$$
\begin{cases}
R_{n+1} = (r+1)R_n - \gamma R_n^2 - bR_nF_n,\\
F_{n+1} = cR_nF_n + (1-d)F_n
\end{cases}
$$

Buscamos las soluciones, así, hagamos:

$$
\begin{cases}
x = (r+1)x - \gamma x^2 - bxy\\
y = cxy + (1-d)y
\end{cases}
$$

Si $x=0$, $y=(1-d)y$. Así $y=0$ ó

$$
1=1-d \iff d=0
$$

Así $(0,0)$ es p. equilibrio.

Si $x\neq 0$,

$$
\begin{cases}
1=(r+1)-\gamma x-by\\
y=cxy+(1-d)y
\end{cases}
$$

$\longrightarrow$ si $y=0$,  
$$
1=(r+1)-\gamma x \iff 0=r-\gamma x
$$

Así
$$
x=\frac{r}{\gamma}
$$

y.e.
$$
\left(\frac{r}{\gamma},0\right)
$$

es p. equilibrio.

$\longrightarrow$ si $y\neq 0$ tenemos que:

$$
1=cx+(1-d), \quad \text{ent. } 0=cx-d
$$

i.e.
$$
x=\frac{d}{c}
$$

y como:

$$
1=(r+1)-\gamma x-by, \quad \text{ent. } 0=r-\gamma x-by
$$

i.e.
$$
by=r-\frac{\gamma d}{c}
$$

y.e.
$$
y=\frac{r}{b}-\frac{\gamma d}{bc}
$$

Así
$$
\left(\frac{d}{c}, \frac{r}{b}-\frac{\gamma d}{bc}\right)
$$

es sol. equilibrio.

**c)** Determina el punto crítico (solución de equilibrio y valor de los parámetros) donde ocurre una bifurcación transcrítica. Interpreta biológicamente esta bifurcación.

Como nuestras soluciones de equilibrio son:

$$(0,0),\ (1,0),\ \left(\frac{d}{c}, \frac{\gamma}{b}-\frac{\gamma d}{bc}\right).$$

Como una bifurcación transcrítica ocurre que 2 soluciones de equilibrio colapsan en una sola solución para ciertos valores de los parámetros. Así en este caso las soluciones $(1,0), \left(\frac{d}{c}, \frac{\gamma}{b}+\frac{\gamma d}{bc}\right)$  
colapsarían, pero $(1,0)$ no tiene parámetros lo que nos dice que $\left(\frac{d}{c}, \frac{\gamma}{b}-\frac{\gamma d}{bc}\right)\to (1,0)$. Así $\frac{d}{c}=1$ i.e. $d=c$. Así $\frac{\gamma}{b}-\frac{\gamma d}{bc}=0$.

ahora veamos que tiene un comportamiento de una bifurcación transcrítica i.e. veamos que para $d<c$ hay una estabilidad y para  
$c<d$ la estabilidad se invierte. Así:  

calculemos la Jacobiana del sistema

$$
\begin{cases}
x=(r+1)x-\gamma x^2-bxy\\
y=cxy+(1-d)y
\end{cases}
$$

$$
J=
\begin{pmatrix}
(r+1)-2\gamma x-by & -bx \\
cy & cx+(1-d)
\end{pmatrix}
$$

Así:

$$
J_{(1,0)}=
\begin{pmatrix}
(r+1)-2\gamma & -b \\
0 & c+(1-d)
\end{pmatrix}
=
\begin{pmatrix}
1-\gamma & -b \\
0 & c+(1-d)
\end{pmatrix}
$$

Como la matriz es triangular superior,  
sus valores propios son la diagonal i.e.

$$
\lambda_1=1-\gamma \quad \text{y} \quad \lambda_2=c+(1-d).
$$
