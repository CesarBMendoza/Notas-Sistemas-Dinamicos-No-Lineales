**Ciclo de Krebs (Glucosa)**

$$
\begin{cases}
\dot{x} = -x + a y + x^2 y, \\
\dot{y} = b - a y - x^2 y,
\end{cases}
$$

Como $x, y$ son concentraciones, supondremos que $x, y \ge 0$ y $a, b > 0$.
Busquemos las coordenadas; lo hacemos:

$$
-x + a y + x^2 y = 0
\quad \Rightarrow \quad
b - a y - x^2 y = 0.
$$

De la primera ecuaciﺃ٣n,

$$
y(a + x^2) = x
\quad \Rightarrow \quad
y_1 = \frac{x}{a + x^2}.
$$

De la segunda ecuaciﺃ٣n,

$$
- y(a + x^2) = -b
\quad \Rightarrow \quad
y_2 = \frac{b}{a + x^2}.
$$

Para $x$ grande,

$$
\dot{x} \approx x^2 y,
$$

$$
\dot{y} \approx -x^2 y.
$$

Hagamos

$$
\frac{\dot{y}}{\dot{x}} \approx -1.
$$

Entonces,

$$
\frac{b - a y - x^2 y}{-x + a y + x^2 y}
= \frac{\dot{y}}{\dot{x}}.
$$

Asﺃﺝ,

$$
\lim_{x \to \infty} \frac{\dot{y}}{\dot{x}}
= \lim_{x \to \infty} \frac{-2 x^2 y}{-1 + 2 x^2 y}
= -1.
$$

$$
\dot{x} + \dot{y}
= -x + a y + x^2 y + b - a y - x^2 y
= -x + b.
$$

Entonces,

$$
\dot{x} + \dot{y} > 0.
$$

Es decir,

$$
\dot{y} > -\dot{x},
$$

i.e.

$$
\frac{\dot{y}}{\dot{x}} \to -1.
$$

Necesitamos que

$$
-x + b < 0.
$$

Asﺃﺝ,

$$
\frac{\dot{y}}{\dot{x}} < -1,
$$

y nos gustarﺃﺝa aplicar el **Teorema de Poincarﺃﺏ**.

Le hacemos un hoyo,

pero para que las soluciones no salgan por la abertura que hicimos.

Concluimos que el punto de equilibrio es inestable.

Y como

$$
-x + a y + x^2 y = 0,
$$

$$
b - a y - x^2 y = 0,
$$

entonces

$$
x - b = 0,
\quad \text{i.e. } x = b.
$$

Por lo tanto,

$$
y = \frac{b}{a + b^2},
$$

y el punto de equilibrio es

$$
\left(b,\; \frac{b}{a + b^2}\right).
$$

El Jacobiano es

$$
J =
\begin{pmatrix}
1 + 2xy & a + x^2 \\
-2xy & -a - x^2
\end{pmatrix}.
$$

Evaluando en el equilibrio $(b,\; b/(a+b^2))$,

$$
J(b,p) =
\begin{pmatrix}
1 + \dfrac{2b^2}{a + b^2} & a + b^2 \\
-2\dfrac{b^2}{a + b^2} & -a - b^2
\end{pmatrix}.
$$

Asﺃﺝ,

$$
\operatorname{tr}(J)
= 1 + \frac{2b^2}{a + b^2} - a - b^2
= \frac{-a - b^2 + 2b^2 - (a + b^2)}{a + b^2}.
$$

$$
\frac{-a + b^2 - a^2 - 2ab^2 - b^4}{a + b^2}.
$$

Y como queremos que

$$
\operatorname{tr}(J) > 0,
$$

entonces

$$
-a + b^2 - a^2 - 2ab^2 - b^4 > 0,
$$

i.e.

$$
b^4 - b^2(1 - 2a) + a^2 + a < 0.
$$

Y el lﺃﺝmite de la regiﺃ٣n se da en

$$
b^4 - b^2(1 - 2a) + a^2 + a = 0.
$$

Asﺃﺝ,

$$
b^2
= \frac{(1 - 2a) \pm \sqrt{(1 - 2a)^2 - 4(a^2 + a)}}{2}.
$$

Pero como no queremos un punto silla, necesitamos que

$$
\det(J) > 0.
$$

Por tanto:

$$
\det
\begin{pmatrix}
-1 + \dfrac{2b^2}{a + b^2} & a + b^2 \\
-2\dfrac{b^2}{a + b^2} & -a - b^2
\end{pmatrix}
= a^2 + b^2 > 0.
$$

Y en la regiﺃ٣n

$$
b^4 - b^2(1 - 2a) + a^2 + a < 0,
$$

se tenemos espirales hacia el punto de equilibrio.
