```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos el sistema
def sistema(x, y):
    dx = y
    dy = y * (1 - 3*x**2 - 2*y**2) - x
    return dx, dy

# Creamos una malla de puntos
x = np.linspace(-2.5, 2.5, 400)
y = np.linspace(-2.5, 2.5, 400)
X, Y = np.meshgrid(x, y)

# Evaluamos el campo de direcciones
DX, DY = sistema(X, Y)

# Normalizamos, evitando divisiones por cero
magnitude = np.sqrt(DX**2 + DY**2)
magnitude[magnitude == 0] = 1  # evita división por cero

DX_norm = DX / magnitude
DY_norm = DY / magnitude

# Graficamos el streamplot
plt.figure(figsize=(8, 8))
plt.streamplot(X, Y, DX_norm, DY_norm, density=1.2, linewidth=1, arrowsize=1.2)
plt.title('Campo de direcciones del sistema')
plt.xlabel('$x$')
plt.ylabel(r'$\dot{x}$')
plt.grid(True)
plt.axis('equal')
plt.show()
```
