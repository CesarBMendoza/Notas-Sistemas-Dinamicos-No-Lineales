**Lotka–Volterra (presa–depredador)**

$$
\dot{y}
= r y\left(1 - \frac{y}{k}\right)
- \frac{a y z}{y + A}
$$

$$
\dot{z}
= s z\left(
\frac{y}{y + A}
- \frac{J}{J + A}
\right)
$$
donde: 

$\displaystyle\frac{a y z}{y + A} = \text{Respuesta funcional del depredador}$
y

$\displaystyle\frac{s y z}{y + A} = \text{Respuesta numérica del depredador}$

Se asume que las presas obedecen a un comportamiento logístico.

$$
\dot{y} = y\left(1 - \frac{y}{30}\right) - \frac{y z}{y + 10},
$$

$$
\dot{z} = z\left(\frac{y}{y + 10} - \frac{3}{5}\right).
$$

notemos que $(0,0)$ es solución de equilibrio.

Pues para $z=0$

$$
y\left(1 - \frac{y}{30}\right) = 0
\;\Rightarrow\;
y = 0 \;\text{o}\; y = 30.
$$

Así, tenemos que $(0,0)$ y $(30,0)$ son soluciónes de equilibrio.

Para $y \neq 0$ y $z \neq 0$ se tiene:

$$
\begin{cases}
1 - \dfrac{y}{30} - \dfrac{z}{y + 10} = 0, \\
\dfrac{y}{y + 10} - \dfrac{3}{5} = 0.
\end{cases}
$$

De la segunda ecuación,

$$
\frac{y}{y + 10} = \frac{3}{5}
\;\Rightarrow\;
y = 15.
$$

Sustituyendo en la primera ecuación,

$$
1 - \frac{15}{30} - \frac{z}{15 + 10} = 0
\;\Rightarrow\;
z = 12.5.
$$

Así, el punto de equilibrio es

$$
(y, z) = (15, 12.5).
$$

Y veremos sus estabilidades con el Jacobiano, el cual es:

$$
J =
\begin{pmatrix}
1 - \dfrac{y}{15} - \dfrac{10z}{(y + 10)^2}
&
-\dfrac{y}{y + 10}
\\[8pt]
\dfrac{10z}{(y + 10)^2}
&
\dfrac{y}{y + 10} - \dfrac{3}{5}
\end{pmatrix}.
$$

Para $(0,0)$, tenemos que:

$$
J_{(0,0)} =
\begin{pmatrix}
1 & 0 \\
0 & -\dfrac{3}{5}
\end{pmatrix}.
$$

Así,

$$
\det(J_{(0,0)}) = -\frac{3}{5}<0
\qquad
\operatorname{tr}(J_{(0,0)}) = \frac{2}{5}.
$$

Por lo tanto, $(0,0)$ es un **punto silla**.

Perop, para $(30,0)$,

$$
J_{(30,0)} =
\begin{pmatrix}
-1 & -\dfrac{3}{4} \\
0 & \dfrac{3}{20}
\end{pmatrix}.
$$

Como

$$
\det(J_{(30,0)}) = -\frac{3}{20}<0
$$

se concluye que $(30,0)$ Tambien es también un **punto silla**.

Por ultimo, Tenemos que:

Para $(15, 12.5)$,

$$
J_{(15,12.5)} =
\begin{pmatrix}
-\dfrac{1}{5} & -\dfrac{3}{5} \\
\dfrac{1}{5} & 0
\end{pmatrix}.
$$

Y como

$$
\det(J_{(15,12.5)})=\frac{3}{25} > 0,
\qquad
\operatorname{tr}(J_{(15,12.5)}) = -\frac{1}{5},
$$

y

$$
(\operatorname{tr} J_{(15,12.5)})^2 - 4\det(J_{(15,12.5)})
= \frac{1}{25} - 4\left(\frac{3}{25}\right) < 0,
$$

se concluye que $(15, 12.5)$ es un **nodo degenerado estable**.
