**Reacciones Oscilantes**

Reacciones químicas oscilantes.

- Yodo y Dioxido de cloro
- Ácido malónico.

Tienen aplicaciones en biología:
- El latido del corazón puede “hacer olas” en un modelo 3D.

$$
\begin{cases}
\dot{x} = a - x - \dfrac{4xy}{1 + x^2}, \\
\dot{y} = b x \left(1 - \dfrac{y}{1 + x^2}\right),
\end{cases}
$$

donde **$x$ = la concentración de yodo** y

**$y$ representa la concentración de $ClO_2$**

Vemos que hay una órbita cerrada en el cuadrante positivo $(x,y) > 0$,
cuando $a, b$ cumplen ciertas condiciones.

Busquemos primero las ceroclinas

$$
0 = a - x - \frac{4xy}{1 + x^2}
\quad \Rightarrow \quad
a - x = \frac{4xy}{1 + x^2}
$$

es decir,

$$
y = \frac{(a - x)(1 + x^2)}{4x}.
$$

Por otro lado,

$$
0 = b x \left(1 - \frac{y}{1 + x^2}\right)
\quad \Rightarrow \quad
y = 1 + x^2.
$$

Tenemos un punto de equilibrio, pero nos gustaría ver que dicho punto es repulsor,
para que las soluciones salgan de él.

Calculamos $J$ y $\det(J) > 0$, pero la estabilidad de dicho punto se basa en $\operatorname{tr}(J)$,
por lo que buscamos cuándo

$$
\operatorname{tr}(J) > 0.
$$`
