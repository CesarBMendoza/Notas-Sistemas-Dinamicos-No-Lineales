Sabemos que $m\ddot{x} = F(x)$  
y $V(x)$ será la energía potencial.

$$
F(x) = -\frac{dV}{dx},
\qquad
m\ddot{x} = -\frac{dV}{dx}.
$$

Entonces,

$$
m\ddot{x} + \frac{dV}{dx} = 0.
$$

Multiplicando por $\dot{x}$,

$$
m\ddot{x}\dot{x} + \dot{x}\frac{dV}{dx} = 0.
$$

Pero como

$$
\dot{x}\frac{dV}{dx} = \frac{dx}{dt}\frac{dV}{dx} = \frac{dV}{dt},
$$

se obtiene

$$
\frac{d}{dt}\left(\frac{1}{2}m\dot{x}^2 + V(x)\right) = 0.
$$

Definimos la energía total como

$$
E(x,\dot{x}) = \frac{1}{2}m\dot{x}^2 + V(x).
$$

Así,

$$
\frac{dE}{dt} = 0,
$$

por tanto $E$ es constante sobre las soluciones $x(t)$.
 
Pero recordemos que, Un sistema con una cantidad conservada $E(x)$,  
tal que $E(x)$ no es constante en ningún subconjunto abierto de $\mathbb{R}^2$, se llama **conservativo**.

Por lo que, A las soluciones que son constantes de $E$ sobre las soluciones $x(t)$ se les conoce como **cantidad conservada**
o simplemente **primera integral**.



# Ejemplo

Consideremos nuevamente

$$
m\ddot{x} = F(x),
$$

con

$$
V(x) = -\frac{1}{2}x^2 + \frac{1}{4}x^4.
$$

Entonces,

$$
F(x) = -V'(x),
$$

por tanto,

$$
m\ddot{x} = x - x^3,
$$

para $m = 1$.  
haciendo $y = \dot{x}$.

Tomamos el sistema

$$
\begin{cases}
\dot{x} = y, \\
\dot{y} = x - x^3.
\end{cases}
$$

y Calculemos las soluciones de equilibrio.

para $y^* = 0$, se tiene que $x^* = 1,\,-1,\,0$

Así, las soluciones de equilibrio son

$$
(0,0), \quad (1,0), \quad (-1,0).
$$

Y el Jacobiano es

$$
J =
\begin{pmatrix}
0 & 1 \\
1 - 3x^2 & 0
\end{pmatrix}.
$$

Por lo que:

$$
J_{(0,0)} =
\begin{pmatrix}
0 & 1 \\
1 & 0
\end{pmatrix}
\quad \Rightarrow \quad
\det(J_{(0,0)}) = -1,
$$

De donde $(0,0)$ es un **punto silla**.

Evaluando en $(1,0)$ (y análogamente en $(-1,0)$),

$$
J_{(1,0)} =
\begin{pmatrix}
0 & 1 \\
-2 & 0
\end{pmatrix}.
$$

Y como

$$
\operatorname{tr}(J_{(1,0)}) = 0
\quad \text{y} \quad
\det(J_{(1,0)}) = 2,
$$

no podemos concluir nada solo con el Jacobiano, Pero:

$$
E(x,y) = \frac{1}{2}m\dot{x}^2 + V(x)
= \frac{1}{2}y^2 - \frac{1}{2}x^2 + \frac{1}{4}x^4.
$$

Y tenemos que

$$
E : \mathbb{R}^2 \to \mathbb{R}.
$$

Por lo que, para $E(x,y) = C$, son curvas de nivel,
y dichas curvas son las soluciones de la ecuación.



**Teorema:**  
En un sistema de ecuaciones diferenciales autónomo conservativo,
con cantidad conservada $E$,
las soluciones de equilibrio correspondientes a mínimos o máximos locales de $E$
son **centros no lineales**.
