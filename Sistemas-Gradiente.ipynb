### Sistemas Gradientes

Sea $V : \mathbb{R}^2 \to \mathbb{R}$ tal que

$$
\dot{x} = \nabla V.
$$

**Teorema:**  
En un sistema gradiente no existen ciclos límite.


ya que:

Supongamos que existe una solución no constante $x(t)$ periódica, es decir,

$$
x(t) = x(t+T).
$$

Entonces

$$
0 = \Delta _T V = V(x(t+T)) - V(x(t))
= \int_t^{t+T} \frac{dV}{d\tau}\, d\tau
= \int_t^{t+T} \nabla V \cdot \dot{x}\, d\tau
= \int_t^{t+T} \|\dot{x}\|^2 \, d\tau.
$$

y esto nos diria que $\dot x = 0$ es decir, $x$ es constante $\blacktriangledown$ pero esto contradice con nuestro supuesto, por lo que no existen soluciones periodicas

# Ejemplo

Consideremos

$$
x + \dot{x} + (\dot{x})^3 = 0
$$

Definimos la energía

$$
E(x,\dot{x}) = \frac{1}{2}\left(x^2 + \dot{x}^2\right)
$$

Si existe una solución periódica pero no constante, entonces

$$
0
= \Delta_T E
= \int_t^{t+T} \frac{dE}{d\tau}\, d\tau
= \int_t^{t+T} \nabla E(x,\dot{x}) \cdot (\dot{x}, \ddot{x}) \, d\tau1
= \int_t^{t+T} \left( \dot{x}\ddot{x} + x\dot{x} \right)\, d\tau
= \int_t^{t+T} \dot{x}(\ddot{x} + x)\, d\tau
$$

Usando la ecuación
$$
\ddot{x} + (\dot{x})^3 + x = 0
$$

tenemos
$$
\ddot{x} + x = -(\dot{x})^3
$$

Por lo tanto,
$$
\Delta_T E
= \int_t^{t+T} -(\dot{x})^4 \, d\tau \le 0
$$
y la igualdad solo se da si
$$
\dot{x} = 0
$$

Por tanto, la ecuación
$$
\ddot{x} + (\dot{x})^3 + x = 0
$$
**no tiene soluciones periódicas no constantes**


**Teorema.**  
Sea $\dot{x} = f(x)$ un sistema, con $x^*$ solución de equilibrio.

Si existe una función $V:\mathbb{R}^2 \to \mathbb{R}$ tal que  
$\forall D \subset \mathbb{R}^2$ conexo con $x^* \in D$ y se cumple:

- $V(x) > 0 \quad \forall x \in D \setminus \{x^*\}$,
- $V(x^*) = 0$,
- $\dot V(x) < 0 \quad \forall x \in D \setminus \{x^*\}$,
- $\dot V(x^*) = 0$,

entonces $x^*$ es **estable** y $D$ es un subconjunto de su **cuenca de atracción**.


# Ejemplo 
Consideremos el sistema
$$
\begin{cases}
\dot x = -x + 4y, \\
\dot y = -x - y^3
\end{cases}
$$

**Función de Lyapunov.**  
Tomamos
$$
V(x,y) = x^2 + 4y^2
$$

Sea
$$
D = \mathbb{R}^2
$$

Notamos que
$$
V(x,y) = 0 \quad \text{si y sólo si} \quad (x,y) = (0,0)
$$

Calculamos la derivada de $V$ a lo largo de las trayectorias:
$$
\dot V = \nabla V \cdot (\dot x, \dot y)
       = (2x,\,8y)\cdot(\dot x,\dot y)
$$

Sustituyendo el sistema:
$$
\dot V = (2x,8y)\cdot(-x+4y,\,-x-y^3)
$$

Desarrollando:
$$
\dot V
= -2x^2 + 8xy - 8xy - 8y^4
= -2x^2 - 8y^4 < 0
$$
para todo $(x,y)\neq(0,0)$

Por lo tanto, $(0,0)$ es un **punto de equilibrio estable** y además
$$
D = \mathbb{R}^2
$$
es la **cuenca de atracción**
