## Teorema (Liénard)

Un sistema de Liénard
$$
\ddot x + g(x) + f(x)\dot x = 0
$$

tiene un **único ciclo límite estable** si se cumplen las siguientes condiciones:

1. $f,g \in C^1$
2. $g(x)$ es impar.
3. $g(x) > 0$ para todo $x > 0$.
4. $f(x)$ es par.
5. $F(x) = \displaystyle \int_0^x f(s)\,ds$ es tal que: Existe un $x_0 > 0$
   - $F(x_0) = 0$,
   - $F(x) < 0$ para $0 < x < x_0$,
   - $F(x) > 0$ para $x > x_0$


## Ejemplo

### Oscilador armónico amortiguado

Consideremos el oscilador armónico amortiguado dado por

$$
m\ddot{x} = -kx - \gamma \dot{x}.
$$

Dividiendo entre $m$, obtenemos

$$
\ddot{x} + \frac{\gamma}{m}\dot{x} + \frac{k}{m}x = 0.
$$

Tomando la forma de Liénard, identificamos:

$$
g(x) = \frac{k}{m}x,
$$

$$
f(x) = \frac{\gamma}{m}.
$$


### Ecuación de Van der Pol

La ecuación de Van der Pol está dada por

$$
\ddot{x} + \mu (x^2 - 1)\dot{x} + x = 0,
$$

donde $\mu > 0$.

Identificando las funciones del sistema de Liénard, tenemos:

$$
g(x) = x,
$$

$$
f(x) = \mu (x^2 - 1).
$$

Además,

$$
F(x) = \int_0^x (u^2 - 1)\,du
$$

entonces

$$
F(x) = \left[\frac{u^3}{3} - u\right]_0^x
     = \frac{x^3}{3} - x.
$$

Para encontrar los ceros de $F(x)$ resolvemos

$$
\frac{x^3}{3} - x = 0,
$$

lo que equivale a

$$
x(x^2 - 3) = 0.
$$

Por tanto,

$$
x = 0 \quad \text{o} \quad x_0 = \pm \sqrt{3}.
$$

En particular, para $x_0 = \sqrt{3}$ se cumple que

$$
F(x_0) = 0.
$$

Ahora, haciendo

$$
\ddot{x} + \mu (x^2 - 1)\dot{x} = -x.
$$

Observamos que

$$
\left(\dot{x} + \mu(\frac{x^3}{3} - x)\right)' = -x,
$$

y por lo tanto

$$
\frac{d}{dt}\left[\dot{x} + \mu F(x)\right] = -x,
$$

donde

$$
F(x) = \frac{x^3}{3} - x.
$$

Definimos la variable auxiliar

$$
w = \dot{x} + \mu F(x).
$$

Entonces

$$
\dot{w} = -x.
$$

Así, el sistema de segundo orden puede escribirse como el sistema equivalente de primer orden

$$
\begin{cases}
\dot{x} = w - \mu F(x), \\
\dot{w} = -x.
\end{cases}
$$

Si además definimos

$$
y = \dfrac{w}{\mu},
$$

entonces

$$
\dot{y} = \dfrac{\dot{w}}{\mu} = -\dfrac{x}{\mu}.
$$
