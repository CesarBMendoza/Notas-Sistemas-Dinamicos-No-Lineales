**Simbiosis**

Modelo:

$$
\dot{y} = y(\lambda - a y + b z) = \lambda y - a y^2 + b y z,
$$

$$
\dot{z} = z(\mu + c y - d z) = \mu z + c y z - d z^2.
$$

como es un modelo biologico, consideraremos que $a,b,c,d>0$

Si la simbiosis es obligatoria, $\lambda, \mu < 0$.


Si es facultativa, $\lambda, \mu > 0$.

Ademas, si 

$$
ad > cb \;\Rightarrow\; \frac{a}{b} > \frac{c}{d},
$$

esto nos dice que el crecimiento de la especie $z$ está acotado.

Y veamos que $(0,0)$ es solución, Pues:

para $y = 0$,

$$
z(\mu - d z) = 0,
\;\Rightarrow\;
z = \frac{\mu}{d} \quad o \quad z=0
$$

por lo tanto $\displaystyle\left(0, \frac{\mu}{d}\right)$ y $(0,0)$ son Puntos de equilibrio.

Por otro lado, para $z = 0$,

$$
y(\lambda - a y) = 0
\;\Rightarrow\;
y = \frac{\lambda}{a}\quad o \quad y=0
$$

por lo tanto $\displaystyle\left(\frac{\lambda}{a}, 0\right)$ tambien es punto de equilibrio


Y si ahora $y, z \neq 0$, entonces

$$
\begin{cases}
\lambda - a y + b z = 0, \\
\mu + c y - d z = 0,
\end{cases}
$$

y la solucion al sistema es:

$$
\left(
\frac{d \lambda + b \mu}{a d - b c},
\frac{c \lambda + a \mu}{a d - b c}
\right).
$$

Por lo tanto, nuestros puntos de equilibrio son:

$\displaystyle\left(0, \frac{\mu}{d}\right)$, $(0,0)$, $\displaystyle\left(\frac{\lambda}{a},0\right)$ y $\displaystyle\left(\frac{d \lambda + b \mu}{a d - b c},\frac{c \lambda + a \mu}{a d - b c}\right)$

Veamos ahora su estabilidad.

El Jacobiano del sistema 
$\begin{cases}
\dot{y} = y(\lambda - a y + b z) \\
\dot{z} = z(\mu + c y - d z)
\end{cases}$ Es:
$\quad
J =
\begin{pmatrix}
\lambda - 2 a y + b z & b y \\
c z & \mu + c y - 2 d z
\end{pmatrix}.
$

Por lo tanto, se tiene que:

Para $(0,0)$,

$$
J_{(0,0)} =
\begin{pmatrix}
\lambda & 0 \\
0 & \mu
\end{pmatrix},
$$

entonces

$$
\det(J) = \lambda \mu
\quad \text{y} \quad
\operatorname{tr}(J) = \lambda + \mu
$$

Para $(y^*, z^*) =\displaystyle\left(\frac{d \lambda + b \mu}{a d - b c},\frac{c \lambda + a \mu}{a d - b c}\right)$

$$
J_{(y^*, z^*)} =
\begin{pmatrix}
- a y^* & b y^* \\
c z^* & - d z^*
\end{pmatrix}
\;\Rightarrow\;
\operatorname{tr}(J) = -a y^* - d z^*
$$

y

$$
\det(J) = a d z^* y^* - c b z^* y^*.
$$

Asi, si: $\lambda,\mu > 0$ tales que $ad>bc$ como:


$\det(J_{(0,0)}) = \lambda \mu>0 \quad y \quad\operatorname{tr}(J_{(0,0)}) = \lambda + \mu>0$
entonces $(0,0)$ es **inestable**

Por otro lado, notemos que como estamos en el supuesto de que $a,b,c,d,\lambda,\mu>0$, entonces $y^*,z^*>0$, por lo que:

$\operatorname{tr}(J_{(y^*,z^*)}) = -a y^* - d z^*<0$ y $\det(J_{(y^*,z^*)}) = a d z^* y^* - c b z^* y^*>0$ es decir:
 $(y^*, z^*)$ es un **nodo estable**.

Haciendo el análisis para $(0, z^*)$ y $(y^*, 0)$, se obtiene que son **inestables**.  
Las soluciones que pasan por dichos puntos tienden a $(y^*, z^*)$.


Ahora, si nuestro modelo es el caso obligatorio, $\lambda, \mu < 0$, y como

$$
J_{(0,0)} =
\begin{pmatrix}
\lambda & 0 \\
0 & \mu
\end{pmatrix},
$$

entonces

$$
\operatorname{tr}(J) = \lambda + \mu < 0
\quad \text{y} \quad
\det(J) > 0.
$$

Así, $(0,0)$ es estable.

Pero es el único punto estable, por lo que las especies tienden a extinguirse.
